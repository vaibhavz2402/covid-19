```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None
```

```python
response = requests.get('https://api.covid19api.com/all')
```

```python
data = response.json()
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.drop(columns = ['CityCode'])
```

```python
df.shape
```

#### Spliting TimeSeries

```python
df['Date'][0].split('T')[0].split('-')
```

```python
df['Year'] = pd.Series()
df['Month'] = pd.Series()
```

```python
end  = df.shape[0]

for i in range(0, end):
    li = []
    li = df['Date'][i].split('T')[0].split('-')
    df['Year'][i] = int(li[0])
    df['Month'][i] = int(li[1])
    df['Date'][i] = int(li[2])
    df['Active'][i] = df['Confirmed'][i] - df['Deaths'][i] - df['Recovered'][i] 
```

```python
df = df.drop(columns = ['City', 'Province'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.Year = df.Year.astype('int')
df.Month = df.Month.astype('int')
```

```python
code = df.CountryCode.unique()
code
```

```python
len(code)
```

```python
data_frame = [*range(0,len(code),1)]     #creating list for storing country wise data 
i = 0
for val in code:
    data_frame[i] = df[df.CountryCode == val]
    i = i+1
```

```python
type(data_frame[100])
```

```python
for frame in range(len(data_frame)):
    data_frame[frame] = data_frame[frame].reset_index(drop = True)
```

##### Countries containing more than 98 values have data for its cities also, So we have to segregate the data city wise for those countries

### Finding countries with city data

```python
big_index = [] 
for frame in range(0,len(data_frame)):
    if(data_frame[frame].shape[0] > 98):
        big_index.append(frame)
```

```python
big_index
```

```python
big_countries = [*range(0,len(big_index),1)]
big_code = [*range(0,len(big_index),1)]

j = 0
for i in big_index:
    big_countries[j] = data_frame[i]
    big_code[j] = code[i]
    j = j+1
```

```python
big_code
```

```python
big_countries[7]
```

```python
code = code.tolist()

for i in big_index:
    del data_frame[i]
    del code[i]
```
